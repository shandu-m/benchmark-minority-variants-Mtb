## Compare H37Rv to L1234

### ISS

```python
fig, axs = plt.subplots(2, 5, figsize=(13, 6))

genomes = ["H37Rv", "L1234"]

for genome_i in range(len(genomes)):

    genome = genomes[genome_i]

    ISS_genome_df = ISS_depth_summary_df[ISS_depth_summary_df.genome == genome]
    
    for depth_i in range(len(depths)):

        depth = depths[depth_i]
    
        ISS_depth_df = ISS_genome_df[ISS_genome_df.expected_depth == depth]
    
        sns.histplot(data=ISS_depth_df, x="value_mean", ax=axs[genome_i][depth_i])

plt.tight_layout()

plt.show()
```

```python
# use Mann-Whitney U test (two-sided)
# - not normally distributed for L1234 samples
# - unequal sample sizes
# - unequal variances
# bonferroni-correction for multiple pairwise tests

all_tools_results_df = pd.DataFrame(columns=["depth", "p1", "p2", "p2-p1", "test_stat", "p_val"])
df_i = 0

for depth in depths:

    depth_df = ISS_depth_summary_df[ISS_depth_summary_df.expected_depth == depth]

    Rv_values = depth_df[depth_df.genome == "H37Rv"]["value_mean"].values
    L1234_values = depth_df[depth_df.genome == "L1234"]["value_mean"].values

    Rv_mean = np.mean(Rv_values)
    L1234_mean = np.mean(L1234_values)

    test_result = mannwu(Rv_values, L1234_values)
    all_tools_results_df.loc[df_i] = [depth, "H37Rv", "L1234", L1234_mean-Rv_mean, test_result[0], test_result[1]]
    df_i += 1

raw_pvalues = all_tools_results_df.p_val.values
reject, corrected_pvals, _, _ = multipletests(raw_pvalues, alpha=0.001, method='bonferroni')

all_tools_results_df["corrected_p_val"] = corrected_pvals
all_tools_results_df["decision"] = reject

all_tools_results_df
```

## Compare replicates

### ISS

```python
simulations = range(1, 6)
simulation_combos = list(combinations(simulations, 2))

all_tools_results_df = pd.DataFrame(columns=["depth", "p1", "p2", "p2-p1", "test_stat", "p_val"])
df_i = 0

for depth in depths:

    depth_df = ISS_depth_summary_df[ISS_depth_summary_df.expected_depth == depth]

    for p1, p2 in simulation_combos:

        p1_values = depth_df[depth_df.simulation == p1]["value_mean"].values
        p2_values = depth_df[depth_df.simulation == p2]["value_mean"].values

        p1_mean = np.mean(p1_values)
        p2_mean = np.mean(p2_values)

        test_result = mannwu(p1_values, p2_values)

        all_tools_results_df.loc[df_i] = [depth, p1, p2, p2_mean-p1_mean, test_result[0], test_result[1]]
        df_i += 1

raw_pvalues = all_tools_results_df.p_val.values
corrected_pvals = false_discovery_control(raw_pvalues, method='bh')

all_tools_results_df["corrected_p_val"] = corrected_pvals
all_tools_results_df["decision"] = (np.array(corrected_pvals) <= 0.001).astype(int)

all_tools_results_df
```

### Compute standard error for each simulator

```python
# standard error of simulated AFs

se_df = pd.DataFrame(columns=["simulator", "AF_under10", "AF_over10"])
df_i = 0

for simulator in simulators:

    # gather ste values for each frequency across all depths
    lower_AF_ste_values, higher_AF_ste_values = [], []

    for freq in freqs:

        freq_df = freq_variant_AFs_by_depth[freq]
        sim_freq_df = freq_df[freq_df.simulator == simulator]
            
        AF_values = sim_freq_df["AF"].values

        freq_ste = sstats.sem(AF_values)

        if freq < 0.1:
            lower_AF_ste_values.append(freq_ste)
        else:
            higher_AF_ste_values.append(freq_ste)

    lower_AF_ste_mean = np.mean(lower_AF_ste_values)
    higher_AF_ste_mean = np.mean(higher_AF_ste_values)
        
    se_df.loc[df_i] = [simulator, lower_AF_ste_mean, higher_AF_ste_mean]
    df_i += 1

se_df
        
```

### Compute average difference between the simulated AF and the expected AF in each region

```python
# standard error of simulated AFs

diff_df = pd.DataFrame(columns=["simulator", "region", "low_AF_average_diff", "high_AF_average_diff"])
df_i = 0

simulator_dicts = {"ISS": ISS_freq_variant_AFs_by_region, "ART": ART_freq_variant_AFs_by_region}

for simulator in simulators:

    freq_variant_AFs_by_region = simulator_dicts[simulator]

    for region in regions:

        low_AF_differences, high_AF_differeces = [], []

        for freq in freqs:

            AF = freq*100

            freq_df = freq_variant_AFs_by_region[freq]
            region_freq_df = freq_df[freq_df.region == region]

            AF_differences = np.array(region_freq_df["AF"].values) - AF

            if freq < 0.1:

                low_AF_differences += list(AF_differences)

            else:
                high_AF_differeces += list(AF_differences)
    
        low_AF_mean = np.mean(low_AF_differences)
        high_AF_mean = np.mean(high_AF_differeces)
            
        diff_df.loc[df_i] = [simulator, region, low_AF_mean, high_AF_mean]
        df_i += 1

diff_df
        
```
