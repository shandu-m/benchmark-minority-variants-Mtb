```python
def x_transform(freq):

    if freq < 1:
        freq = freq*100

    if freq < 10:
        return freq*10
    else:
        return 0.5*freq + 50
```

```python
x_transformed_points = []
for freq in freqs:
    x_transformed_points.append(x_transform(freq))
x_transformed_points
```

# [done] H37Rv: Tool performance by depth for increasing AF

## Precision

```python
fig, axs = plt.subplots(3, 2, sharex=True, sharey=True, figsize=(11, 10))

for tool_i in range(len(tools)):

    tool = tools[tool_i]
    tool_name = tool_mapping[tool]

    tool_df = pd.read_csv(f"./data/accuracy_by_AF_and_depth/ISS_H37Rv_{tool}_metrics_by_depth_AF.csv")
    tool_df["plot_x"] = tool_df["min_freq"].apply(x_transform)

    ax_i, ax_j = tool_i//2, tool_i%2
    ax = axs[ax_i][ax_j]

    ax.grid(True, linewidth=0.8, zorder=0)
    
    ax.set_ylim(0, 1.05)
    
    ax.set_title(tool_name, fontsize=18)
        
    sns.lineplot(tool_df, x="plot_x", y="precision", hue="depth", style="depth",
                 markers=True, markersize=14, palette=depth_colors, zorder=2, legend=False, ax=ax)
    
    ax.tick_params(labelsize=18)
    
    x_ticks = x_transformed_points
    ax.set_xticks(x_ticks)
    ax.set_xticklabels([str(int(x*100))+"%" for x in freqs], fontsize=16, rotation=45, horizontalalignment="center")

    ax.set_ylabel("")
    ax.set_xlabel("")
                                                         
fig.text(0.5, -0.025, "Simulated variant allele frequency", ha='center', fontsize=20)
fig.text(-0.035, 0.5, "Average precision", rotation='vertical', va='center', fontsize=20) 
    
plt.tight_layout()
    
file_name = "./figures/accuracy_by_AF_and_depth/ISS_H37Rv_overall_precision_by_AF_and_depth.jpeg"
plt.savefig(file_name, bbox_inches='tight', dpi = 300 , transparent = True)

file_name = "./figures/accuracy_by_AF_and_depth/ISS_H37Rv_overall_precision_by_AF_and_depth.pdf"
plt.savefig(file_name, bbox_inches='tight', dpi = 300 , transparent = True)

plt.show()
```

# [done] H37Rv + L1-4: F1 per region for all AF

## All depths

```python
DR_precision_per_AF_summary_df = pd.DataFrame(columns=["min_freq"] + tools)
DR_precision_per_AF_summary_df["min_freq"] = list(np.array(freqs)*100)
DR_precision_per_AF_summary_df.set_index("min_freq", inplace=True)

DR_recall_per_AF_summary_df = pd.DataFrame(columns=["min_freq"] + tools)
DR_recall_per_AF_summary_df["min_freq"] = list(np.array(freqs)*100)
DR_recall_per_AF_summary_df.set_index("min_freq", inplace=True)

DR_df = pd.read_csv(f"./data/accuracy_by_AF_and_depth/ISS_allAF_metrics_by_AF_DR.csv")

for tool in tools:

    tool_DR_df = DR_df[DR_df.tool == tool]

    for freq in freqs:

        freq_tool_DR_df = tool_DR_df[tool_DR_df.min_freq == freq]
        
        avg_freq_tool_DR_precision = freq_tool_DR_df.precision.mean()
        avg_freq_tool_DR_recall = freq_tool_DR_df.recall.mean()

        freq = freq*100

        DR_precision_per_AF_summary_df.loc[freq, tool] = avg_freq_tool_DR_precision
        DR_recall_per_AF_summary_df.loc[freq, tool] = avg_freq_tool_DR_recall

print("DR PRECISION")
display(DR_precision_per_AF_summary_df)
print("DR RECALL")
display(DR_recall_per_AF_summary_df)
```
