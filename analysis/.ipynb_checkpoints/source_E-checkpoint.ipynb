# [done] AF differences for different replicates

## H37Rv

````python
Rv_AF_diffs_by_simulation = {t:{s:[] for s in range(1, 6)} for t in tools}

count = 0
for tag in tagsRv:
    
    count += 1
    if count % 10 == 0:
        print(count, end=" ")
        
    sim_i = int(tag.split("_")[0].strip("sim"))
    
    pileup_AF_df = pd.read_csv("{0}/{1}.AF.csv".format(Rv_mpileup_source, tag))
    
    tag_summary_df = pd.read_csv("{0}/{1}.variant_summary.csv".format(ISS_Rv_mutant_summary_source, tag), low_memory=False).set_index(["POS", "ALT"])
    tag_summary_df = tag_summary_df[tag_summary_df.GT == 1]
        
    for i in pileup_AF_df.index:
        
        POS, ALT = pileup_AF_df.loc[i, "POS"], pileup_AF_df.loc[i, "ALT"]
        pileup_AF = pileup_AF_df.loc[i, "AF"]
        
        for tool in tools:
            
            tool_AF = tag_summary_df.loc[(POS, ALT), "{}_AF".format(tool)]
            
            if not np.isnan(tool_AF):
                
                tool_AF_diff = (tool_AF - pileup_AF)*100
                
                Rv_AF_diffs_by_simulation[tool][sim_i].append(tool_AF_diff)
                    
````

````python
with open("./data/AF_accuracy/ISS_H37Rv_AF_differences_by_tool_simulation_dict.pkl", "wb") as out_f:
    pickle.dump(Rv_AF_diffs_by_simulation, out_f)
````

````python
VD_outliers_df = pd.DataFrame(columns=["POS", "ALT", "PILEUP_AF"]+["{}_AF".format(t) for t in tools]+["INFO"])
df_i = 0

count = 0
for tag in tagsRv:
    
    count += 1
    if count % 10 == 0:
        print(count, end=" ")
    
    pileup_AF_df = pd.read_csv("{0}/{1}.AF.csv".format(Rv_mpileup_source, tag))
    
    tag_summary_df = pd.read_csv("{0}/{1}.variant_summary.csv".format(ISS_Rv_mutant_summary_source, tag), low_memory=False).set_index(["POS", "ALT"])
    tag_summary_df = tag_summary_df[tag_summary_df.GT == 1]
        
    for i in pileup_AF_df.index:
        
        POS, ALT = pileup_AF_df.loc[i, "POS"], pileup_AF_df.loc[i, "ALT"]
        pileup_AF = pileup_AF_df.loc[i, "AF"]
        
        for tool in tools:
            
            tool_AF = tag_summary_df.loc[(POS, ALT), "{}_AF".format(tool)]
            
            if tool == "VD" and not np.isnan(tool_AF):
                
                tool_AF_diff = (tool_AF - pileup_AF)*100
                
                if tool_AF_diff < -10:
                    
                    VD_outliers_df.loc[df_i, ["POS", "ALT", "PILEUP_AF"]] = [POS, ALT, pileup_AF*100]
                    
                    for t in tools:      
                        try:
                            VD_outliers_df.loc[df_i, "{}_AF".format(t)] = tag_summary_df.loc[(POS, ALT), "{}_AF".format(t)]*100
                        except ValueError:
                            VD_outliers_df.loc[df_i, "{}_AF".format(t)] = np.nan
                            
                    VD_outliers_df.loc[df_i, "INFO"] = [tag]
                            
                    df_i += 1

for tag in tagsL1234:
    
    count += 1
    if count % 10 == 0:
        print(count, end=" ")
    
    pileup_AF_df = pd.read_csv("{0}/{1}.AF.csv".format(L1234_mpileup_source, tag))
    
    tag_summary_df = pd.read_csv("{0}/{1}.variant_summary.csv".format(ISS_L1234_mutant_summary_source, tag), low_memory=False).set_index(["POS", "ALT"])
    tag_summary_df = tag_summary_df[tag_summary_df.GT == 1]
        
    for i in pileup_AF_df.index:
        
        POS, ALT = pileup_AF_df.loc[i, "POS"], pileup_AF_df.loc[i, "ALT"]
        pileup_AF = pileup_AF_df.loc[i, "AF"]
        
        for tool in tools:
            
            tool_AF = tag_summary_df.loc[(POS, ALT), "{}_AF".format(tool)]
            if type(tool_AF) == pd.Series:
                tool_AF = tool_AF.values[0]
            
            if tool == "VD" and not np.isnan(tool_AF):
                
                tool_AF_diff = (tool_AF - pileup_AF)*100
                
                if tool_AF_diff < -10:
                    
                    VD_outliers_df.loc[df_i, ["POS", "ALT", "PILEUP_AF"]] = [POS, ALT, pileup_AF*100]
                    
                    for t in tools:      
                        try:
                            VD_outliers_df.loc[df_i, "{}_AF".format(t)] = tag_summary_df.loc[(POS, ALT), "{}_AF".format(t)]*100
                        except ValueError:
                            VD_outliers_df.loc[df_i, "{}_AF".format(t)] = np.nan
                            
                    VD_outliers_df.loc[df_i, "INFO"] = [tag]
                            
                    df_i += 1
                    
````
