# Mapping quality

```python
with open(f"./data/FP_characteristics/ISS_L1234_MQ_distribution.pkl", "rb") as in_f:
    ISSL1234_MQ_distr = pickle.load(in_f)
```

```python
FB_TP_MQ = np.array(ISSL1234_MQ_distr["FB"]["TP"])
FB_FP_MQ = np.array(ISSL1234_MQ_distr["FB"]["FP"])
```

```python
sum(pd.isna(FB_TP_MQ)), sum(pd.isna(FB_FP_MQ))
```

```python
# percentage of TP variants at which the mapping quality is the max mapping quality
FB_TP_MQ_distr = np.array(FB_TP_MQ)
print(f"max MQ = {np.max(FB_TP_MQ_distr)}")
print((sum(FB_TP_MQ_distr == np.max(FB_TP_MQ_distr))/len(FB_TP_MQ_distr))*100)
```

```python
fig, ax = plt.subplots(figsize=(5,4))

ax.grid(axis="y", linewidth=0.3, zorder=0)

vp = sns.violinplot([FB_TP_MQ, FB_FP_MQ], inner="quart", fill=True, palette=[TP_col, FP_col], ax=ax)

faded_cols = [faded_P_cols["TP"], faded_P_cols["FP"]]
for pc_i in range(len(vp.collections)):
    pc = vp.collections[pc_i]
    base_color = faded_cols[pc_i]
    pc.set_facecolor(base_color)
    pc.set_edgecolor("black")
    pc.set_linewidth(1)

ax.set_xticks([0, 1])
ax.set_xticklabels(["True positives", "False positives"])

ax.tick_params(labelsize=12)
ax.set_ylabel("Mapping quality at variant site", fontsize=14)

plt.tight_layout()

file_name = "./figures/FB_FP_filtering/ISS_L1234_FB_MQ_distribution.jpeg"
plt.savefig(file_name, bbox_inches='tight', dpi = 300 , transparent = True)

file_name = "./figures/FB_FP_filtering/ISS_L1234_FB_MQ_distribution.pdf"
plt.savefig(file_name, bbox_inches='tight', dpi = 300 , transparent = True)

plt.show()
```
