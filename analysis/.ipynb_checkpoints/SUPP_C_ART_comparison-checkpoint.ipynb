```python
weighted_F1_dfs = {"ISS": ISS_Rv_weighted_F1_df,
                   "ART": ART_Rv_weighted_F1_df}
```

```python
weighted_F1_avg_df = pd.DataFrame(columns=["Tool", "Simulation", "weighted_F1_avg", "sim_tool"])
df_i = 0

for sim_tool in sim_tools:

    simtool_df = weighted_F1_dfs[sim_tool]

    for tool in tools:
        
        tool_simtool_df = simtool_df[simtool_df.tool == tool]
        
        for simulation in range(1, 6):
            
            sim_tool_simtool_df = tool_simtool_df[tool_simtool_df.simulation == simulation]
            
            weighted_F1_avg = sim_tool_simtool_df.weighted_F1.mean()
            
            weighted_F1_avg_df.loc[df_i] = [tool_mapping[tool], simulation, weighted_F1_avg, sim_tool]
            df_i += 1

weighted_F1_avg_df
```

```python
order_df = pd.DataFrame(columns=["tool", "avg_F1"])
df_i = 0

weighted_F1_avg_df_to_order = weighted_F1_avg_df[weighted_F1_avg_df.sim_tool == "ART"]

for tool in tools:
    
    tool_name = tool_mapping[tool]
    
    tool_df = weighted_F1_avg_df_to_order[weighted_F1_avg_df_to_order.Tool == tool_name]
    
    avg_F1 = tool_df.weighted_F1_avg.mean()
    
    order_df.loc[df_i] = [tool_name, avg_F1]
    df_i += 1
    
order_df.sort_values("avg_F1", inplace=True, ascending=False)
tool_order = list(order_df.tool.values)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
    
ax.set_ylim(0.3, 1.01)

ax.grid(axis="y", linewidth=0.8, zorder=0)

ax.yaxis.set_minor_locator(MultipleLocator(0.1))

## ISS ##
plot_df = weighted_F1_avg_df[weighted_F1_avg_df.sim_tool == "ISS"]

sns.stripplot(data=plot_df, x="Tool", y="weighted_F1_avg", jitter=0.3, color="white", order=tool_order,
              ax=ax, size=20, marker=sim_tool_markers["ISS"], edgecolor="black", linewidth=2, alpha=0.75)
##

## ART ##
plot_df = weighted_F1_avg_df[weighted_F1_avg_df.sim_tool == "ART"]

sns.stripplot(data=plot_df, x="Tool", y="weighted_F1_avg", jitter=0.3, hue="Tool", order=tool_order,
              palette=tool_colors_, ax=ax, size=25, marker=sim_tool_markers["ART"], edgecolor="black", linewidth=1, alpha=0.8)
##

ax.tick_params(axis='y', labelsize=20) 
ax.yaxis.set_minor_locator(MultipleLocator(0.1))

ax.tick_params(axis='x', labelrotation=45, labelsize=20) 
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticklabels(), fontdict={"horizontalalignment": "right"})

ax.set_xlabel("")
ax.set_ylabel("Average weighted F1", fontsize=22, labelpad=22)  

plt.tight_layout()

file_name = "./figures/ART_comparison/ISS_vs_ART_weighted_F1.jpeg"
plt.savefig(file_name, bbox_inches='tight', dpi = 300 , transparent = True)

file_name = "./figures/ART_comparison/ISS_vs_ART_weighted_F1.pdf"
plt.savefig(file_name, bbox_inches='tight', dpi = 300 , transparent = True)

plt.show()
```

```python
weighted_F1_dfs["ISS"].head()
```

```python
# use Mann-Whitney U test (two-sided)
# - not normally distributed
# - unequal sample sizes
# - unequal variances
# FDR correction for multiple pairwise tests

all_tools_results_df = pd.DataFrame(columns=["tool", "p1", "p2", "p2-p1", "test_stat", "p_val"])
df_i = 0

p1, p2 = "ISS", "ART"

for tool in tools:

    ISS_tool_df = weighted_F1_dfs["ISS"]
    ART_tool_df = weighted_F1_dfs["ART"]

    ISS_values = ISS_tool_df.weighted_F1.values
    ART_values = ART_tool_df.weighted_F1.values

    ISS_mean = np.mean(ISS_values)
    ART_mean = np.mean(ART_values)

    test_result = mannwu(ISS_values, ART_values)
    all_tools_results_df.loc[df_i] = [tool, p1, p2, ART_mean-ISS_mean, test_result[0], test_result[1]]
    df_i += 1

raw_pvalues = all_tools_results_df.p_val.values
corrected_pvals = false_discovery_control(raw_pvalues, method='bh')

all_tools_results_df["corrected_p_val"] = corrected_pvals
all_tools_results_df["decision"] = (np.array(corrected_pvals) <= 0.05).astype(int)

all_tools_results_df
```

```python
F1_region_plot_df = pd.DataFrame(columns=["region", "simulation", "Tool", "F1_avg", "sim_tool"])
df_i = 0

for sim_tool in sim_tools:

    Rv_precision_recall_df = Rv_precision_recall_dfs[sim_tool]
    L1234_precision_recall_df = L1234_precision_recall_dfs[sim_tool]

    for region in regions:
            
        region_df = Rv_precision_recall_df[Rv_precision_recall_df.region == region]
        
        for sim_i in range(1, 6):
            
            sim_region_df = region_df[region_df.simulation == sim_i]
        
            for tool in tools:
    
                tool_sim_region_df = sim_region_df[sim_region_df.tool == tool]
                                    
                p_avg = tool_sim_region_df.precision.mean()
                r_avg = tool_sim_region_df.recall.mean()

                F1_avg = calc_F1(p_avg, r_avg)

                F1_region_plot_df.loc[df_i] = [region, sim_i, tool_mapping[tool], F1_avg, sim_tool]
                df_i += 1
            
        
F1_region_plot_df
```

```python
tool_order = {r:[] for r in regions}

for region in regions:

    order_df = pd.DataFrame(columns=["tool", "avg_F1"])
    df_i = 0

    region_df = F1_region_plot_df[(F1_region_plot_df.region == region) & (F1_region_plot_df.sim_tool == "ART")]
    
    for tool in tools:
        
        tool_name = tool_mapping[tool]
        
        tool_region_df = region_df[region_df.Tool == tool_name]

        avg_F1 = tool_region_df.F1_avg.mean()

        order_df.loc[df_i] = [tool_name, avg_F1]
        df_i += 1
    
    order_df.sort_values("avg_F1", inplace=True, ascending=False)
    tool_order[region] = list(order_df.tool.values)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(14, 4), sharey=True)

for region_i in range(len(regions)):

    ax = axs[region_i]
    
    region = regions[region_i]
    region_name = region_mapping[region]

    ax.set_title(region_name, fontsize=18)
    
    region_tool_order = tool_order[region]
    
    region_plot_df = F1_region_plot_df[F1_region_plot_df.region == region]

    ax.grid(axis='y', which='both', linewidth=0.8, zorder=1)
    
    ax.set_ylim(-0.05, 1.01)

    ISS_region_plot_df = region_plot_df[region_plot_df.sim_tool == "ISS"]
    
    sns.stripplot(data=ISS_region_plot_df, x="Tool", y="F1_avg", jitter=0.3, color="white",
                  order=region_tool_order, size=15, marker=sim_tool_markers["ISS"], edgecolor="black", linewidth=1, alpha=0.75,
                  ax=ax, zorder=1)

    ART_region_plot_df = region_plot_df[region_plot_df.sim_tool == "ART"]
    
    sns.stripplot(data=ART_region_plot_df, x="Tool", y="F1_avg", jitter=0.3, hue="Tool", palette=tool_colors_,
                  order=region_tool_order, size=17, marker=sim_tool_markers["ART"], edgecolor="black", linewidth=1, alpha=0.8,
                  ax=ax, zorder=2)
    
    ax.tick_params(axis='y', labelsize=18) 
    
    ax.tick_params(axis='x', labelrotation=45, labelsize=18) 
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(ax.get_xticklabels(), fontdict={"horizontalalignment": "right"})
        
    ax.yaxis.set_minor_locator(MultipleLocator(0.1))
    ax.yaxis.set_major_locator(MultipleLocator(0.2))

    ax.set_xlabel("")
    ax.set_ylabel("")
        
    ax.set_ylabel("Average F1", va='center', rotation='vertical', fontsize=20, labelpad=20)
 
plt.tight_layout()

file_name = "./figures/ART_comparison/ISS_vs_ART_H37Rv_F1_by_region.jpeg"
plt.savefig(file_name, bbox_inches='tight', dpi = 300 , transparent = True)

file_name = "./figures/ART_comparison/ISS_vs_ART_H37Rv_F1_by_region.pdf"
plt.savefig(file_name, bbox_inches='tight', dpi = 300 , transparent = True)

plt.show()
```

```python
for region in regions:
    
    print(">", region)
    region_df = pd.DataFrame(columns=["F1_diff_(ART-ISS)", "tool"])
    df_i = 0

    ISS_region_df = ISS_Rv_precision_recall_by_region_df[ISS_Rv_precision_recall_by_region_df.region == region]
    ART_region_df = ART_Rv_precision_recall_by_region_df[ART_Rv_precision_recall_by_region_df.region == region]
                
    for tool in tools:

        ISS_region_tool_precision_lst = ISS_region_df[ISS_region_df.tool == tool].precision.values
        ISS_region_tool_recall_lst = ISS_region_df[ISS_region_df.tool == tool].recall.values
        ISS_region_tool_F1_lst = []
        for (p, r) in zip(ISS_region_tool_precision_lst, ISS_region_tool_recall_lst):
            ISS_region_tool_F1_lst.append(calc_F1(p, r))
        ISS_region_tool_F1_lst = np.array(ISS_region_tool_F1_lst)

        ART_region_tool_precision_lst = ART_region_df[ART_region_df.tool == tool].precision.values
        ART_region_tool_recall_lst = ART_region_df[ART_region_df.tool == tool].recall.values
        ART_region_tool_F1_lst = []
        for (p, r) in zip(ART_region_tool_precision_lst, ART_region_tool_recall_lst):
            ART_region_tool_F1_lst.append(calc_F1(p, r))
        ART_region_tool_F1_lst = np.array(ART_region_tool_F1_lst)

        F1_diff_lst = ART_region_tool_F1_lst - ISS_region_tool_F1_lst
        avg_F1_diff = np.mean(F1_diff_lst)

        region_df.loc[df_i] = [avg_F1_diff, tool]
        df_i += 1
        
    display(region_df)
    print("Mean diff = {}".format(region_df.iloc[:, 0].mean()))
    print()
        
```

```python
# use Mann-Whitney U test (two-sided)
# - not normally distributed
# - equal sample sizes
# - unequal variances
# FDR correction for multiple pairwise tests

plot_region_combos = list(combinations(plot_regions, 2))

all_tools_results_df = pd.DataFrame(columns=["region1", "region2", "region2-region1_mean_diff", "p_val"])
df_i = 0

for region1, region2 in plot_region_combos:

    region1_values = L1234_region_FPR_df[L1234_region_FPR_df.region == region1]["FPR"].values
    region2_values = L1234_region_FPR_df[L1234_region_FPR_df.region == region2]["FPR"].values
    
    test_result = mannwu(region1_values, region2_values)

    all_tools_results_df.loc[df_i] = [region1, region2, np.mean(region2_values)-np.mean(region1_values), test_result[1]]
    df_i += 1

raw_pvalues = all_tools_results_df.p_val.values
corrected_pvals = false_discovery_control(raw_pvalues, method='bh')

all_tools_results_df["corrected_p_val"] = corrected_pvals
all_tools_results_df["decision"] = (np.array(corrected_pvals) <= 0.05).astype(int)

all_tools_results_df
```

```python
# use Mann-Whitney U test (two-sided)
# - not normally distributed
# - equal sample sizes
# - unequal variances
# FDR correction for multiple pairwise tests

plot_region_combos = list(combinations(plot_regions, 2))

all_tools_results_df = pd.DataFrame(columns=["region1", "region2", "region2-region1_mean_diff", "p_val"])
df_i = 0

for region1, region2 in plot_region_combos:

    region1_values = Rv_region_FPR_df[Rv_region_FPR_df.region == region1]["FPR"].values
    region2_values = Rv_region_FPR_df[Rv_region_FPR_df.region == region2]["FPR"].values
    
    test_result = mannwu(region1_values, region2_values)

    all_tools_results_df.loc[df_i] = [region1, region2, np.mean(region2_values)-np.mean(region1_values), test_result[1]]
    df_i += 1

raw_pvalues = all_tools_results_df.p_val.values
corrected_pvals = false_discovery_control(raw_pvalues, method='bh')

all_tools_results_df["corrected_p_val"] = corrected_pvals
all_tools_results_df["decision"] = (np.array(corrected_pvals) <= 0.05).astype(int)

all_tools_results_df
```
