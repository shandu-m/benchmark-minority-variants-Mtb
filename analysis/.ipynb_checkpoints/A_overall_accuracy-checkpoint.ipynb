```python
# test for statistical significance in pairwise rankings

# use Mann-Whitney U test (two-sided)
# - not normally distributed
# - equal sample sizes
# - unequal variances
# bonferroni-correction for multiple pairwise tests

all_tools_results_df = pd.DataFrame(columns=["tool", "metric", "avg", "raw_pval"])
final_df_i = 0

# L1234 DR
region = "DR"

ranking_df = pd.DataFrame(columns=["tool", "weighted_F1_avg"])
df_i = 0

for tool in tools:

    tool_weighted_F1_df = Rv_weighted_F1_df[Rv_weighted_F1_df.tool == tool]

    weighted_F1_average = tool_weighted_F1_df.weighted_F1.mean()
    ranking_df.loc[df_i] = [tool, weighted_F1_average]
    df_i += 1

ranking_df.sort_values("weighted_F1_avg", ascending=False, inplace=True)
ranking_df.reset_index(drop=True, inplace=True)

ranking_order = ranking_df.tool.values

raw_pvals = []
for tool_1_i in range(len(ranking_order)-1):
    
    tool_1 = ranking_order[tool_1_i]
    tool_2 = ranking_order[tool_1_i + 1]

    tool_1_values = Rv_weighted_F1_df[Rv_weighted_F1_df.tool == tool_1]["weighted_F1"].values
    tool_2_values = Rv_weighted_F1_df[Rv_weighted_F1_df.tool == tool_2]["weighted_F1"].values

    test_result = mannwu(tool_1_values, tool_2_values)
    raw_pvals.append(test_result[1])

raw_pvals.append(np.nan)

for i in ranking_df.index:

    all_tools_results_df.loc[final_df_i] = [ranking_df.loc[i, "tool"], "weighted_F1", ranking_df.loc[i, "weighted_F1_avg"], raw_pvals[i]]
    final_df_i += 1

raw_p_value_ix = []
raw_pvalues = []
for i in all_tools_results_df.index:
    raw_p = all_tools_results_df.loc[i, "raw_pval"]
    if not np.isnan(raw_p):
        raw_p_value_ix.append(i)
        raw_pvalues.append(raw_p)
        
#reject, corrected_pvals, _, _ = multipletests(raw_pvalues, alpha=0.001, method='bonferroni')

corrected_pvals = false_discovery_control(raw_pvalues, method='bh')

all_tools_results_df["corrected_p_val"] = [np.nan]*all_tools_results_df.shape[0]
all_tools_results_df["decision"] = [np.nan]*all_tools_results_df.shape[0]

for add_ix in range(len(raw_p_value_ix)):
    ix = raw_p_value_ix[add_ix]
    all_tools_results_df.loc[ix, "corrected_p_val"] = corrected_pvals[add_ix]
    all_tools_results_df.loc[ix, "decision"] = (corrected_pvals[add_ix] <= 0.05).astype(int)

all_tools_results_df
```

## Difference between DR H37Rv and DR L1234

```python
region = "DR"

Rv_region_df = Rv_precision_recall_by_region_df[Rv_precision_recall_by_region_df.region == region]
L1234_region_df = L1234_DR_precision_recall_df[L1234_DR_precision_recall_df.region == region]

tool_DR_F1_values = pd.DataFrame(columns=["tool", "genome", "DR_F1"])
df_i = 0

for tool in tools:

    tool_name = tool_mapping[tool]

    Rv_tool_region_df = Rv_region_df[Rv_region_df.tool == tool]
    precision_values = np.array(Rv_tool_region_df.precision.values)
    recall_values = np.array(Rv_tool_region_df.recall.values)
    
    for P, R in zip(precision_values, recall_values):
        tool_DR_F1_values.loc[df_i] = [tool, "H37Rv", calc_F1(P, R)]
        df_i += 1

    L1234_tool_region_df = L1234_region_df[L1234_region_df.tool == tool]
    precision_values = np.array(L1234_tool_region_df.precision.values)
    recall_values = np.array(L1234_tool_region_df.recall.values)
    
    for P, R in zip(precision_values, recall_values):
        tool_DR_F1_values.loc[df_i] = [tool, "L1234", calc_F1(P, R)]
        df_i += 1
```

```python
# DR statistical test for H37Rv vs L1234 (tag-wise for each tool)

for tool in tools:

    tool_df = tool_DR_F1_values[tool_DR_F1_values.tool == tool]

    Rv_tool_values = tool_df[tool_df.genome == "H37Rv"]["DR_F1"].values
    L1234_tool_values = tool_df[tool_df.genome == "L1234"]["DR_F1"].values

    print(tool)
    print(np.mean(Rv_tool_values), np.mean(L1234_tool_values))
    print(np.var(Rv_tool_values), np.var(L1234_tool_values))

    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.hist(Rv_tool_values, color="lightblue", edgecolor="black")
    ax2.hist(L1234_tool_values, color="lightblue", edgecolor="black")
    plt.show()
    plt.close()
```

```python
# use Mann-Whitney U test (two-sided)
# - not normally distributed
# - unequal sample sizes
# - unequal variances
# bonferroni-correction for multiple pairwise tests

all_tools_results_df = pd.DataFrame(columns=["tool", "avg_pooled_F1", "p1", "p2", "p2-p1", "test_stat", "p_val"])
df_i = 0

for tool in tools:

    tool_F1_df = tool_DR_F1_values[tool_DR_F1_values.tool == tool]

    Rv_values = tool_F1_df[tool_F1_df.genome == "H37Rv"]["DR_F1"].values
    L1234_values = tool_F1_df[tool_F1_df.genome == "L1234"]["DR_F1"].values

    all_values = list(Rv_values) + list(L1234_values)

    avg_pooled_F1 = np.mean(all_values)

    Rv_F1_mean = np.mean(Rv_values)
    L1234_F1_mean = np.mean(L1234_values)

    test_result = mannwu(Rv_values, L1234_values)
    all_tools_results_df.loc[df_i] = [tool, avg_pooled_F1, "H37Rv", "L1234", L1234_F1_mean-Rv_F1_mean, test_result[0], test_result[1]]
    df_i += 1

raw_pvalues = all_tools_results_df.p_val.values
#reject, corrected_pvals, _, _ = multipletests(raw_pvalues, alpha=0.001, method='bonferroni')
corrected_pvals = false_discovery_control(raw_pvalues, method='bh')

all_tools_results_df["corrected_p_val"] = corrected_pvals

all_tools_results_df
```

## Determine difference in DR F1 score between L1234 samples

```python
# statistical test lineage differences amongst L1-4 (tag-wise for each tool)

tool_F1_dfs = {}

for tool in tools:

    tool_F1_df = pd.DataFrame(columns=["F1", "lineage"])
    df_i = 0

    tool_DR_precision_recall_df = L1234_DR_precision_recall_df[L1234_DR_precision_recall_df.tool == tool]
    
    for lineage in range(1, 5):

        lineage_DR_precision_recall_df = tool_DR_precision_recall_df[tool_DR_precision_recall_df.lineage == lineage]

        lineage_precision_values = np.array(lineage_DR_precision_recall_df.precision.values)
        lineage_recall_values = np.array(lineage_DR_precision_recall_df.recall.values)

        for P, R in zip(lineage_precision_values, lineage_recall_values):
            lineage_F1 = calc_F1(P, R)
            tool_F1_df.loc[df_i] = [lineage_F1, lineage]
            df_i += 1

    tool_F1_dfs[tool] = tool_F1_df
```

```python
for tool in tools:

    tool_F1_df = tool_F1_dfs[tool]

    fig, axs = plt.subplots(1, 4)

    for lineage in range(1, 5):

        ax = axs[lineage-1]

        ax.hist(tool_F1_df[tool_F1_df.lineage == lineage]["F1"].values, color="lightblue", edgecolor="black")

    plt.show()
    plt.close()
```

```python
# use Mann-Whitney U test (two-sided)
# - not normally distributed
# - equal sample sizes
# - unequal variances
# bonferroni-correction for multiple pairwise tests

lineages = range(1, 5)
lineage_combos = list(combinations(lineages, 2))

all_tools_results_df = pd.DataFrame(columns=["tool", "p1", "p2", "p2-p1", "test_stat", "p_val"])
df_i = 0

for tool in tools:

    tool_F1_df = tool_F1_dfs[tool]

    for p1, p2 in lineage_combos:

        p1_tool_F1_values = tool_F1_df[tool_F1_df.lineage == p1]["F1"].values
        p2_tool_F1_values = tool_F1_df[tool_F1_df.lineage == p2]["F1"].values

        p1_F1_mean = np.mean(p1_tool_F1_values)
        p2_F1_mean = np.mean(p2_tool_F1_values)

        test_result = mannwu(p1_tool_F1_values, p2_tool_F1_values)

        all_tools_results_df.loc[df_i] = [tool, p1, p2, p2_F1_mean-p1_F1_mean, test_result[0], test_result[1]]
        df_i += 1

raw_pvalues = all_tools_results_df.p_val.values
#reject, corrected_pvals, _, _ = multipletests(raw_pvalues, alpha=0.001, method='bonferroni')
corrected_pvals = false_discovery_control(raw_pvalues, method='bh')

all_tools_results_df["corrected_p_val"] = corrected_pvals
all_tools_results_df["decision"] = (np.array(corrected_pvals) <= 0.05).astype(int)

all_tools_results_df
```

```python
# test for statistical significance in pairwise rankings

# use Mann-Whitney U test (two-sided)
# - not normally distributed
# - equal sample sizes
# - unequal variances
# bonferroni-correction for multiple pairwise tests

all_tools_results_df = pd.DataFrame(columns=["region", "genome", "tool", "metric", "avg", "raw_pval"])
final_df_i = 0

# H37Rv all regions
for region in regions:

    region_name = region_mapping[region]

    region_df = Rv_precision_recall_by_region_df[Rv_precision_recall_by_region_df.region == region]
    
    ranking_df = pd.DataFrame(columns=["tool", "precision_avg"])
    df_i = 0

    for tool in tools:

        tool_region_df = region_df[region_df.tool == tool]

        precision_average = tool_region_df.precision.mean()
        ranking_df.loc[df_i] = [tool, precision_average]
        df_i += 1

    ranking_df.sort_values("precision_avg", ascending=False, inplace=True)
    ranking_df.reset_index(drop=True, inplace=True)

    ranking_order = ranking_df.tool.values

    raw_pvals = []
    for tool_1_i in range(len(ranking_order)-1):
        
        tool_1 = ranking_order[tool_1_i]
        tool_2 = ranking_order[tool_1_i + 1]

        tool_1_values = region_df[region_df.tool == tool_1]["precision"].values
        tool_2_values = region_df[region_df.tool == tool_2]["precision"].values
    
        test_result = mannwu(tool_1_values, tool_2_values)
        raw_pvals.append(test_result[1])

    raw_pvals.append(np.nan)

    for i in ranking_df.index:

        all_tools_results_df.loc[final_df_i] = [region, "H37Rv", ranking_df.loc[i, "tool"], "precision", ranking_df.loc[i, "precision_avg"], raw_pvals[i]]
        final_df_i += 1

# L1234 DR
region = "DR"

ranking_df = pd.DataFrame(columns=["tool", "precision_avg"])
df_i = 0

for tool in tools:

    tool_L1234_DR_df = L1234_DR_precision_recall_df[L1234_DR_precision_recall_df.tool == tool]

    precision_average = tool_L1234_DR_df.precision.mean()
    ranking_df.loc[df_i] = [tool, precision_average]
    df_i += 1

ranking_df.sort_values("precision_avg", ascending=False, inplace=True)
ranking_df.reset_index(drop=True, inplace=True)

ranking_order = ranking_df.tool.values

raw_pvals = []
for tool_1_i in range(len(ranking_order)-1):
    
    tool_1 = ranking_order[tool_1_i]
    tool_2 = ranking_order[tool_1_i + 1]

    tool_1_values = L1234_DR_precision_recall_df[L1234_DR_precision_recall_df.tool == tool_1]["precision"].values
    tool_2_values = L1234_DR_precision_recall_df[L1234_DR_precision_recall_df.tool == tool_2]["precision"].values

    test_result = mannwu(tool_1_values, tool_2_values)
    raw_pvals.append(test_result[1])

raw_pvals.append(np.nan)

for i in ranking_df.index:

    all_tools_results_df.loc[final_df_i] = [region, "L1234", ranking_df.loc[i, "tool"], "precision", ranking_df.loc[i, "precision_avg"], raw_pvals[i]]
    final_df_i += 1

raw_p_value_ix = []
raw_pvalues = []
for i in all_tools_results_df.index:
    raw_p = all_tools_results_df.loc[i, "raw_pval"]
    if not np.isnan(raw_p):
        raw_p_value_ix.append(i)
        raw_pvalues.append(raw_p)
        
#reject, corrected_pvals, _, _ = multipletests(raw_pvalues, alpha=0.001, method='bonferroni')

corrected_pvals = false_discovery_control(raw_pvalues, method='bh')

all_tools_results_df["corrected_p_val"] = [np.nan]*all_tools_results_df.shape[0]
all_tools_results_df["decision"] = [np.nan]*all_tools_results_df.shape[0]

for add_ix in range(len(raw_p_value_ix)):
    ix = raw_p_value_ix[add_ix]
    all_tools_results_df.loc[ix, "corrected_p_val"] = corrected_pvals[add_ix]
    all_tools_results_df.loc[ix, "decision"] = (corrected_pvals[add_ix] <= 0.05).astype(int)

all_tools_results_df
```

```python
# use Mann-Whitney U test (two-sided)
# - not normally distributed
# - equal sample sizes
# - unequal variances
# bonferroni-correction for multiple pairwise tests

simulations = range(1, 6)
simulation_combos = list(combinations(simulations, 2))

all_tools_results_df = pd.DataFrame(columns=["tool", "p1", "p2", "p2-p1", "test_stat", "p_val"])
df_i = 0

for tool in tools:

    tool_df = Rv_precision_recall_by_region_df[Rv_precision_recall_by_region_df.tool == tool]

    for p1, p2 in simulation_combos:

        p1_tool_F1_values = tool_df[tool_df.simulation == p1]["precision"].values
        p2_tool_F1_values = tool_df[tool_df.simulation == p2]["precision"].values

        p1_F1_mean = np.mean(p1_tool_F1_values)
        p2_F1_mean = np.mean(p2_tool_F1_values)

        test_result = mannwu(p1_tool_F1_values, p2_tool_F1_values)

        all_tools_results_df.loc[df_i] = [tool, p1, p2, p2_F1_mean-p1_F1_mean, test_result[0], test_result[1]]
        df_i += 1

raw_pvalues = all_tools_results_df.p_val.values
reject, corrected_pvals, _, _ = multipletests(raw_pvalues, alpha=0.001, method='bonferroni')

all_tools_results_df["corrected_p_val"] = corrected_pvals
all_tools_results_df["decision"] = reject

all_tools_results_df
```

```python
# use Mann-Whitney U test (two-sided)
# - not normally distributed
# - equal sample sizes
# - unequal variances
# bonferroni-correction for multiple pairwise tests

simulations = range(1, 5)
simulation_combos = list(combinations(simulations, 2))

all_tools_results_df = pd.DataFrame(columns=["tool", "p1", "p2", "p2-p1", "test_stat", "p_val"])
df_i = 0

for tool in tools:

    tool_df = L1234_DR_precision_recall_df[L1234_DR_precision_recall_df.tool == tool]

    for p1, p2 in simulation_combos:

        p1_tool_F1_values = tool_df[tool_df.simulation == p1]["precision"].values
        p2_tool_F1_values = tool_df[tool_df.simulation == p2]["precision"].values

        p1_F1_mean = np.mean(p1_tool_F1_values)
        p2_F1_mean = np.mean(p2_tool_F1_values)

        test_result = mannwu(p1_tool_F1_values, p2_tool_F1_values)

        all_tools_results_df.loc[df_i] = [tool, p1, p2, p2_F1_mean-p1_F1_mean, test_result[0], test_result[1]]
        df_i += 1

raw_pvalues = all_tools_results_df.p_val.values
reject, corrected_pvals, _, _ = multipletests(raw_pvalues, alpha=0.001, method='bonferroni')

all_tools_results_df["corrected_p_val"] = corrected_pvals
all_tools_results_df["decision"] = reject

all_tools_results_df
```
