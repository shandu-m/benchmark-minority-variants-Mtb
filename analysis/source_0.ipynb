# Make pilon VCF tag maps to extract DP and BQ metrics

Pilon includes an entry for each position

## ISS Rv

````python
tag_df = pd.DataFrame(columns=["tag", "path"])
df_i = 0

for depth in depths:
    
    depth_source_dir = ISS_Rv_wgs_source.format(depth)
    
    for mutant_num in range(1, 11):
        for freq in freqs:
            for sim_i in range(1, 6):
                
                tag = "sim{}_h37rv_mutant_{}_{}_{}".format(sim_i, mutant_num, depth, freq)
                
                pilon_vcf = "{0}/{1}/pilon/{1}.vcf.gz".format(depth_source_dir, tag)
                assert(os.path.isfile(pilon_vcf))
                
                tag_df.loc[df_i] = [tag, pilon_vcf]
                df_i += 1

tag_df.to_csv("/n/scratch/users/s/sm624/ISS_H37Rv_simulated_depths/tag_map.csv", index=False, header=False)
````

## Combine

````python
ISS_depth_summary_df = pd.DataFrame(columns=["tag", "genome", "simulation", "expected_depth", "value_mean"])

# H37Rv
ISS_Rv_depth_summary_df = pd.read_csv("/n/scratch/users/s/sm624/ISS_H37Rv_simulated_depths/simulated_depths_summary.csv")
ISS_depth_summary_df["tag"] = list(ISS_Rv_depth_summary_df.tag.values)
ISS_depth_summary_df["genome"] = ["H37Rv"]*ISS_Rv_depth_summary_df.shape[0]
ISS_depth_summary_df["simulation"] = [int(tag.split("_")[0].strip("sim")) for tag in ISS_Rv_depth_summary_df.tag.values]
ISS_depth_summary_df["expected_depth"] = [int(tag.split("_")[-2]) for tag in ISS_Rv_depth_summary_df.tag.values]
ISS_depth_summary_df["value_mean"] = list(ISS_Rv_depth_summary_df.value_mean.values)

# L1234
ISS_L1234_depth_summary_df = pd.read_csv("/n/scratch/users/s/sm624/ISS_L1234_simulated_depths/simulated_depths_summary.csv")
add_df = pd.DataFrame(columns=["tag", "genome", "simulation", "expected_depth", "value_mean"])
add_df["tag"] = list(ISS_L1234_depth_summary_df.tag.values)
add_df["genome"] = ["L1234"]*ISS_L1234_depth_summary_df.shape[0]
add_df["simulation"] = [int(tag.split("_")[0].strip("sim")) for tag in ISS_L1234_depth_summary_df.tag.values]
add_df["expected_depth"] = [int(tag.split("_")[-2]) for tag in ISS_L1234_depth_summary_df.tag.values]
add_df["value_mean"] = list(ISS_L1234_depth_summary_df.value_mean.values)

ISS_depth_summary_df = pd.concat([ISS_depth_summary_df, add_df])

ISS_depth_summary_df.head()
````

````python
ISS_depth_summary_df.to_csv("./data/simulated_data_statistics/ISS_depth_summary.csv", index=False)
````

## By region

### ISS Rv

````python
for freq in freqs:
    
    raw_read_counts_df = pd.DataFrame(columns=["depth", "AF", "region"])

    for depth in depths:

        for sim_i in range(1, 6):

            for mutant_num in range(1, 11):

                tag = "sim{}_h37rv_mutant_{}_{}_{}".format(sim_i, mutant_num, depth, freq)

                simulated_mutations_df = pd.read_csv("../../../../mutant_summaries_final/mutant_{}.csv".format(mutant_num)).set_index(["POS", "ALT"])
                simulated_mutations_df.sort_index(inplace=True)
                simulated_mutations_df = simulated_mutations_df[["TYPE"]]
                
                pileup_df = pd.read_csv("{0}/{1}.AF.csv".format(ISS_Rv_pileup_source_dir, tag)).set_index(["POS", "ALT"])
                pileup_df.sort_index(inplace=True)
                
                merged_df = pileup_df.join(simulated_mutations_df, how='left')
                
                read_count_values = np.array(merged_df["AF"].values)*100
                
                add_df = pd.DataFrame(columns=["depth", "AF", "region"])
                add_df["depth"] = [depth]*len(read_count_values)
                add_df["AF"] = list(read_count_values)
                add_df["region"] = merged_df.TYPE.values

                raw_read_counts_df = pd.concat([raw_read_counts_df, add_df])

    raw_read_counts_df.to_csv("./data/simulated_data_statistics/ISS_H37Rv_raw_read_counts_by_region_{}.csv".format(freq), index=False)
````
